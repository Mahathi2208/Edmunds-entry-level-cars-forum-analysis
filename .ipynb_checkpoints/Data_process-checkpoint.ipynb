```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv(r"comments.csv").drop(columns=['Unnamed: 0'])
```

```python
word_freq_map = {}

for comment in data['comments']:
    if type(comment) == float:
         continue
  
    for word in comment.split():
        
        word = re.sub(r'[^\w]', ' ', word).strip().lower()
        
        if " " in word:
            for w in word.split():
                w = re.sub(r'[^\w]', ' ', w).strip().lower()
                if w in word_freq_map:
                    word_freq_map[w] += 1
                else:
                    word_freq_map[w] = 1
        else:
            if word in word_freq_map:
                word_freq_map[word] += 1
            else:
                word_freq_map[word] = 1
  
    

```

```python
word_freq_map

sorted_map = sorted(word_freq_map.items(), key = lambda item: -item[1])
```

```python
sorted_map
```

```python
type(sorted_map)
```

```python
key_word = [x[0] for x in sorted_map]
freq = [x[1] for x in sorted_map]

freq_df = pd.DataFrame({'word': key_word, 'frequency': freq})
```

```python
freq_df
```

```python
freq_df.to_csv('word_freq.csv')
```

```python
model_brand_list = pd.read_csv('car models and brands .csv')
```

```python
model_brand_list
```

```python
model_to_brand = {}

for idx, row in model_brand_list.iterrows():
    model_to_brand[row['Model']] = row['Brand']
```

```python
def isBrand(car, model_list):
    if car in model_list:
        return True
    else:
        return False
```

```python

word_freq_map_model_brand = {}

model_list = list(set([x for x in word_freq_map_model_brand]))

for comment in data['comments']:
    if type(comment) == float:
         continue
    
    cur_user_car_set = set()
    
    for word in comment.split():
        
        word = re.sub(r'[^\w]', ' ', word).strip().lower()
        
        #Make sure each brand mentioned by a user only count as one mention
        if isBrand(word, model_list) and word in cur_user_car_set:
            continue
        
        if " " in word:
            for w in word.split():
                w = re.sub(r'[^\w]', ' ', w).strip().lower()
                
                if isBrand(w, model_list) and w in cur_user_car_set:
                    continue
                
                if w in model_to_brand:
                    w = model_to_brand[w]
                if w in word_freq_map:
                    word_freq_map[w] += 1
                else:
                    word_freq_map[w] = 1
        
        else:
            
            if word in model_to_brand:
                word = model_to_brand[word]
            
            if word in word_freq_map_model_brand:
                word_freq_map_model_brand[word] += 1
            else:
                word_freq_map_model_brand[word] = 1

  
```

```python
new_list = sorted(word_freq_map_model_brand.items(), key = lambda x: -x[1])
new_key_word = [x[0] for x in new_list]
new_freq = [x[1] for x in new_list]
new_freq_df = pd.DataFrame({'word': new_key_word, 'frequency': new_freq})
```
